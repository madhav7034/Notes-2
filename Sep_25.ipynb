```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip = pd.read_excel('/content/tip-amount.xlsx')
```

```python
tip
```

```python
corr = tip.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corr, vmin = .2, vmax = 1, annot=True, linewidths=0.2,cmap='YlGnBu')
```

```python
y = tip['Observed tip amount(yi)']
x = tip.drop(['Observed tip amount(yi)'],axis = 1)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42, test_size = .25)

from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
x_test
```

```python
pred
```

```python
new_data_point = pd.DataFrame({'Meal':[4,7],'Observed total bill(xi)':[90,150]})
pred_amt = model.predict(new_data_point)
pred_amt
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is: \n',mean_squared_error(y_test,pred))
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
data = fetch_california_housing(as_frame=True)
data
```

```python
df = pd.concat([data.data,data.target],axis = 1)
df
```

```python
#OR

df1 = data.frame
df1
```

```python
df1.isna().sum()
```

```python
corr = df1.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corr, vmin = .2, vmax = 1, annot=True, linewidths=0.2,cmap='YlGnBu')
```

```python
num_cols = df1.select_dtypes(include='float64')
num_cols
```

```python
for i in num_cols:
  plt.figure(figsize=(5,5))
  plt.boxplot(df1[i])
  plt.title(i)
```

```python
from scipy.stats.mstats import winsorize
df1['AveRooms'] = winsorize(df1['AveRooms'], limits = (0.05, 0.05))
df1['AveBedrms'] = winsorize(df1['AveBedrms'], limits = (0.10, 0.10))
df1['Population'] = winsorize(df1['Population'], limits = (0.10, 0.10))
df1['AveOccup'] = winsorize(df1['AveOccup'], limits = (0.05, 0.05))
df1['MedInc'] = winsorize(df1['MedInc'], limits = (0.05, 0.05))
```

```python
for i in num_cols:
  plt.figure(figsize=(5,5))
  plt.boxplot(df1[i])
  plt.title(i)
```

```python
  plt.boxplot(df1['AveRooms'])
```

```python
y = df1['MedHouseVal']
x = df1.drop(['MedHouseVal'],axis = 1)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42, test_size = .25)

from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is: \n',mean_squared_error(y_test,pred))
```

```python
from sklearn.metrics import r2_score
print('R2 is: \n',r2_score(y_test,pred))
```
