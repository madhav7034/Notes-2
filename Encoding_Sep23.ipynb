```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
import pandas as pd
```

```python
studentdf = pd.DataFrame({
    'Student ID': [1,2,3,4,5],
    'Performance_Level': ['Low','Medium','High','Medium','Low'],
    'Gender': ['Male','Female','Male','Male','Female'],
    'Math_Score': [75,85,90,78,80],
    'English_Score': [82,88,95,79,85],
    'Major': ['Science','Arts','Science','Commerce','Arts']
    })
studentdf
```

```python
ordinal_enc = OrdinalEncoder(categories=[['Low','Medium','High']])
studentdf['Performance_Level_ord'] = ordinal_enc.fit_transform(studentdf[['Performance_Level']])
```

```python
studentdf
```

```python
label_enc = LabelEncoder()
studentdf['Gender_enc'] = label_enc.fit_transform(studentdf['Gender'])
```

```python
studentdf
```

```python
new_studentdf = pd.get_dummies(studentdf,
                               columns = ['Major'],
                               prefix = 'ma',
                               drop_first=True,
                               dtype='int64')
```
