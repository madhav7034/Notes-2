```python
import requests
url = 'https://api.spacexdata.com/v4/launches'

response1 = requests.get(url)
data1 = response1.json()
data1

rocket_df = pd.DataFrame(data1)
rocket_df.head()
```

```python
rocket = rocket_df[['name','date_utc','success','details','rocket','launchpad']]
rocket
```

```python
rocket.rename(columns={'name':'mission_name', 'rocket':'rocket_id'},inplace=True)
```

```python
import requests
url = 'https://api.spacexdata.com/v4/rockets'

response2 = requests.get(url)
data2 = response2.json()
data2

rock_df = pd.DataFrame(data2)
rock_df.head()
```

```python
rock = rock_df[['name','id','company','country','cost_per_launch']]
rock
```

```python
final = rocket.merge(rock,left_on = 'rocket_id',right_on = 'id',how = 'left')
final
```
